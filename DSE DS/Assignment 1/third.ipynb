Linear Regression - Real Estate Price

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
```

```python
data=pd.read_csv('real_estate_price.csv')
```

```python
data.head()
```

```python
x=data['price']
y=data['size']
plt.scatter(x,y)
plt.xlabel('Price',fontsize=20) 
plt.ylabel('Size',fontsize=20) 
plt.show()
```

```python
x1=sm.add_constant(x) # this adds the y-intercept
results=sm.OLS(y,x1).fit() # Ordinary Least squares model for y on x1
results.summary()
```

```python
plt.scatter(x,y) 
yhat=0.0033*x-122.3349 
fig=plt.plot(x,yhat,lw=4,c='orange',label='regression line') 
plt.xlabel('Price',fontsize=20) 
plt.ylabel('Size',fontsize=20) 
plt.show()
```
